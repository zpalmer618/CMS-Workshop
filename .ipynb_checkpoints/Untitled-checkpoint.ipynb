## Plotting Lesson

```python
import numpy
import os
```

```python
distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
print(distances)
```

```python
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)
```

```python
import matplotlib.pyplot as plt #alias for a long library name
```

```python
plt.plot(data[:,0],data[:,1])
```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
fig_1 = plt.plot(data[:,0],data[:,1], label=sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
plt.figure()
for col in range(1, len(headers)):
    fig = plt.plot(data[:,0],data[:,col], label=headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.legend()

plt.savefig(F'Distances Graph.png')
```

```python
plt.figure() #Create a new figure
for col in range(1, len(headers)): #Starts a new loop to go through each line
    figure = plt.figure() #For every line in the loop it creates a new figure
    fig = plt.plot(data[:,0],data[:,col], label=headers[col]) #plots the new figures for each line of data
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstroms)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png', dpi=300)
```
